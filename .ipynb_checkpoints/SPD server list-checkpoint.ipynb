```python
xls = pd.ExcelFile('SPD_Express_Global - Domain_20200406.xlsx')
filter_files = [col for col in xls.sheet_names if 'Server List' in col]
i=0
list_df = []
for file in filter_files:
    df = pd.read_excel(io='SPD_Express_Global - Domain_20200406.xlsx', sheet_name=file, skiprows=8)
    list_df.append(df)    
```

```python
all =pd.concat(list_df, axis=0)
all.to_csv('allservers.csv', index=False)
all
```
